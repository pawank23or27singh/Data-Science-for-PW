Normalization

```python
import seaborn as sns
import numpy as np
import pandas as pd
```

```python
df=sns.load_dataset('tips')
df.head()
```

```python
from sklearn.preprocessing import MinMaxScaler
min_max=MinMaxScaler()
min_max.fit(df[['total_bill','tip']])
```

```python
min_max.transform(df[['total_bill','tip']])
```

```python
min_max.fit_transform(df[['total_bill','tip']])
```

```python
min_max.transform([[1,3]])
```

```python
import seaborn  as sns
df=sns.load_dataset('iris')
df.head()
```

```python
from sklearn.preprocessing import normalize
df.columns
```

```python
normalize(df[['sepal_length', 'sepal_width', 'petal_length', 'petal_width']])
```

```python
import pandas as pd
pd.DataFrame(normalize(df[['sepal_length', 'sepal_width', 'petal_length', 'petal_width']]),columns=['sepal_length', 'sepal_width', 'petal_length', 'petal_width'])
```
